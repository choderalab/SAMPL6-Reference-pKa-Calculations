```python
from titrato.sampl import SAMPL6DataProvider
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
np.set_printoptions(threshold=9001)
```

```python
loader = SAMPL6DataProvider("Epik/typeII-raw-microscopic.csv", "typeii", "Epik-free-energy")
```

```python
for molecule in range(1,25):
    molid= f"SM{molecule:02d}"
    print(molid)
    curve = loader.load(molid)
    fig,ax = plt.subplots(1,1)
    fig.tight_layout()
    plt.title(molid, fontsize=18)
    plt.ylabel("Free energy (KT)", fontsize=18)
    plt.xlabel("pH", fontsize=18)
    for idx, state in enumerate(curve.state_ids):
        ax.scatter(curve.ph_values, curve.free_energies[idx])
#         plt.ylim([-0.01, 10.01])
    ax2 = plt.twinx()
    ax2.set_ylabel("Number of states in output", fontsize=18)
    plt.plot(curve.ph_values,np.count_nonzero(~np.isinf(curve.free_energies),axis=0), color='k')
```
